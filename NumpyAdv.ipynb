Binary Cross Entropy

Binary Cross Entropy

```python
def binary_cross_entropy(actual, predicted, epsilon=1e-15):
  # Clip predicted values to avoid log(0) and log(1) issues
  predicted = np.clip(predicted, epsilon, 1 - epsilon)
  return -np.mean(actual * np.log(predicted) + (1 - actual) * np.log(1 - predicted))

# Example usage:
# Generate some random actual and predicted values (between 0 and 1 for binary classification)
actual_bce = np.array([1, 0, 1, 1, 0])
predicted_bce = np.array([0.9, 0.1, 0.8, 0.7, 0.2])

print(f"Actual values: {actual_bce}")
print(f"Predicted values: {predicted_bce}")
print(f"Binary Cross Entropy: {binary_cross_entropy(actual_bce, predicted_bce)}")
```
